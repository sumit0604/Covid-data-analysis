```python
#again reading recently saved file and store it in a data vaiable.now we are analysing on coviddata.csv file
data=pd.read_csv("coviddata.csv")
data
```

```python
data= data.iloc[:,1:]
data
```

```python
#for data visualization i am using a matplotlib 

import matplotlib.pyplot as plt
```

```python
#we are going to find the null value in our dataset.....
#basically isnull() fucntion isused to find the null values..
#data.isnull().sum(axis=0).sort_values()/len(data)*100..this will find the percentage of null value.....
#round function is used to get the rounded value and 2 is used to get the 2 decimal value.
round(data.isnull().sum(axis=0).sort_values(ascending=False)/len(data)*100,2)
```

```python
#here we are going to find cases came in each month....that is why we splited our date column in 3 columns date,month,years..
#in this we are using groupby function and passing the month aregument and after that a list where we give our Num cases column
#at last we will do sum of all the cses by sum()
data.groupby('month')['Num Cases'].sum()
```

```python
#here we will analyse the Deceased people month wise....
data[data["Current Status"]=="Deceased"].groupby("month")["Num Cases"].sum()
```

```python
#before this we have to delete those columns who has maximun null values//
# so what i did is.... i have deleted Detected district, Detected city and Status Change Date columns
# by using del data['column name']
# del data['Detected City']
#now removing all null values by dropna()
clean=data.dropna()
# clean.isnull()
# data
clean
```

```python
# here we are doing the sum of those cases whose Current Status is Hospitalized....
graphclean=clean[clean["Current Status"]=="Hospitalized"].groupby("month")["Num Cases"].sum()
graphclean
```

```python
# by plot.bar we are making a bar chart to visualize it graphically
graphclean.plot.bar()
plt.show()
```

```python
# For finding the number of cases of only male month wise
malegraph=clean[clean["Gender"]=="M"].groupby("month")["Num Cases"].sum()
```
